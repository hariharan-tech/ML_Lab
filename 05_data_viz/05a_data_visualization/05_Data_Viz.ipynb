## Data Visualization
Visualizations shown: Plot, Stem, Histogram, Scatter, Box/Whisker, Violin, Pie chart, Heat map, Dist plot, Joint plot, Line plot, LM plot, Pair plot, Count plot, Bar plot, Ridgeline plot

```python
# importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use(['science','no-latex'])

# Project configurations
figSize = (10,4)
```

```python
# Available Datasets in Seaborn
print(f"Available Databases in seaborn : \n{sns.get_dataset_names()}")

# Database directory
database_dir = "./seaborn-data/"

# Loading a simple database
# Equivalent to sns.load_dataset()
dataset_iris = pd.read_csv(f"{database_dir}/iris.csv")

print(f"\n{dataset_iris.info()}")
```

```python
def set_values(ax,x_label,y_label,title):
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    ax.set_title(title)
```

### Plot, Stem Histogram and Scatter functions
1. Plot using matplotlib.pyplot.plot function
    - Used to plot the array using a simple line plot
2. Stem using matplotlib.pyplot.stem function
    - Used to plot the array in a discrete form
3. Histogram using matplotlib.pyplot.hist function
    - Used to plot the distribution of occurence of values in the array
4. Scatter using matplotlib.pyplot.scatter function
    - Used to Scatter the values (instead of plot)

```python
x = dataset_iris["sepal_width"]
y = dataset_iris["sepal_length"]
fig, ax = plt.subplots(2,2, figsize=figSize)
ax[0,0].plot(y,"r")
set_values(ax[0,0],"Index","Sepal Length","Sepal length - Plot")
ax[0,1].stem(x[:100])
set_values(ax[0,1],"Index","Sepal Width","Sepal Width - Stem")
ax[1,0].hist(y)
set_values(ax[1,0],"Index","Sepal Length","Sepal length - Histogram")
ax[1,1].scatter(dataset_iris["sepal_length"],dataset_iris["petal_width"],c="k")
set_values(ax[1,1],"Sepal Length","Sepal WidtH","Sepal Length vs Petal Width - Scatter")
plt.tight_layout()
plt.show()
plt.close("all")
```

### Box/ Whisker and Violin Plot
- Boxplot using seaborn.boxplot function
    - Used to plot the distribution (tells about the minimum and maximum values)
- Violin Plot using seaborn.violin function
    - Same as boxplot but shows the distribution

```python
fig, ax = plt.subplots(1,2,figsize=figSize)
sns.boxplot(data=dataset_iris,x="species",y="sepal_length",ax=ax[0])
set_values(ax[0],"Species","Sepal Length","Species vs Sepal Length - Box Plot")
sns.violinplot(data=dataset_iris,x="species",y="sepal_length",ax=ax[1])
set_values(ax[1],"Species","Sepal Length","Species vs Sepal Length - Violin Plot")
plt.show()
plt.close("all")
```

### Pie Chart
Using matplotlib.pyplot.pie function

```python
species = dataset_iris["species"]
species_count = np.array(species.value_counts())
species = list(set(dataset_iris["species"]))
fig,ax = plt.subplots(1,1,figsize=figSize)
ax.pie(species_count,labels=species)
ax.set_title("Pie Chart for species count")
plt.show()
plt.close("all")
```

### Heat Map
Using seaborn.heatmap

```python
fig, ax = plt.subplots(1,1,figsize=(10,6))
sns.heatmap(dataset_iris.drop(["species"],axis=1)[0:20], annot=True, annot_kws={"size": 7},ax=ax)
plt.show()
plt.close("all")
```

### Dist Plot
Using seaborn.displot function

```python
sns.displot(dataset_iris["sepal_width"])
plt.show()
plt.close("all")
```

### Joint Plot
Using seaborn.jointplot function

```python
sns.jointplot(data=dataset_iris,x="sepal_length",y="sepal_width",kind="kde")
plt.show()
plt.close("all")
```

### Line plot
Using seaborn.lineplot Function

```python
fig, ax = plt.subplots(1,1,figsize=figSize)
sns.lineplot(data=dataset_iris,x="sepal_width",y="sepal_length",ax=ax)
set_values(ax,"Year","Passengers","Count of passengers versus year - Line Plot")
plt.show(); plt.close("all")
```

### LM plot
Using seaborn.lmplot function
- Combines Regression fit plot on the data

```python
sns.lmplot(data=dataset_iris,x="sepal_length",y="sepal_width")
set_values(ax,"Sepal Length","Sepal Width","Sepal Length vs Sepal Width\n- LM Plot (to find the line of regression)")
plt.show()
plt.close("all")
```

### Pair plot
Using seaborn.pairplot function
- To plot multiple pairwise bivariate distributions in a dataset

```python
sns.pairplot(data=dataset_iris)
plt.show()
plt.close("all")
```

### Count plot
Using seaborn.countplot Function

```python
fig, ax = plt.subplots(1,1,figsize=figSize)
dataset_iris["sepal_len5.5"] = dataset_iris["sepal_length"]<5.5
sns.countplot(data=dataset_iris,x="species",hue="sepal_len5.5",ax=ax)
set_values(ax,"Species","Count","Sepal Length less than 5.5 for 3 species - Count Plot")
# plt.show()
# plt.close("all")
```

### Bar plot
Using seaborn.barplot function

```python
fig, ax = plt.subplots(1,1,figsize=figSize)
sns.barplot(data=dataset_iris,x="species",y="sepal_length",ax=ax)
set_values(ax,"Species","Sepal Length","Sepal Length vs Species - Bar plot")
plt.show()
plt.close("all")
```

### Ridgeline plot
- Combined form of Density plots for columns in a dataset
- There is no specific function that does this functionality
- This function is implemented here

```python
import warnings
warnings.filterwarnings("ignore",category=UserWarning)
fig, ax = plt.subplots(4,1,figsize=figSize,sharex=True)
# for i in range(1,5):
#     sns.distplot(np.random.normal(i,2*i,100),ax=ax[i-1])
sns.distplot(dataset_iris["sepal_length"],ax=ax[0])
sns.distplot(dataset_iris["sepal_width"],ax=ax[1])
sns.distplot(dataset_iris["petal_length"],ax=ax[2])
sns.distplot(dataset_iris["petal_width"],ax=ax[3])
plt.show()
plt.close("all")
```

### Overlay Plot
A simple overlay plot is plotted here using matplotlib.pyplot.plt to plot the sepal length and width

```python
fig, ax = plt.subplots(1,1,figsize=(16,5))
ax.plot(dataset_iris["sepal_width"],"r")
ax.plot(dataset_iris["sepal_length"],"b")
ax.legend(["Sepal Width","Sepal Length"])
set_values(ax,"Index","Length (in cm)","Sepal Length and Width - Plot")
plt.show()
plt.close("all")
```
