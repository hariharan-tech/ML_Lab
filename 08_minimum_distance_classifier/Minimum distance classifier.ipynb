```python
import Min_Dis_classifier as mdc
import numpy as np
import matplotlib.pyplot as plt
# plt.style.use(["science","no-latex"])
```

```python
def plot_dist(x,y,z,**kwargs):
    fig = plt.figure(figsize=(7,14))
    ax = fig.add_subplot(2, 1, 1, projection='3d')
    ax.plot_surface(x,y,z,cmap="gnuplot")
    ax = fig.add_subplot(2, 1, 2)
    ax.contourf(x,y,z,40,cmap="gnuplot")
    if len(kwargs.items())!=0:
        if kwargs["flag"]:
            print(kwargs["test_point"])
            ax.scatter(kwargs["test_point"][0],kwargs["test_point"][1],c="white",edgecolors="black")
    # plt.savefig("Variables_uncorrelated.png",dpi=500)
    plt.show()
```

```python
x1 = np.linspace(-10,10,100)
x2 = np.linspace(-10,10,100)
ranges = (x1,x2)
num_of_c = 2
covar_matrix = np.identity(num_of_c)
```

```python
guassians = list()

g1 = mdc.Guassian(np.array([0,0]),covar_matrix,num_of_c,ranges)
g2 = mdc.Guassian(np.array([4,4]),covar_matrix,num_of_c,ranges)

guassians = [g1,g2]

(z1,x1,y1) = g1.iter_norm_dis()
(z2,x2,y2) = g2.iter_norm_dis()
plot_dist(x1,y1,z1+z2)
```

```python
test_pt = np.array([2,0])
distances = list()

for i in guassians:
    distances.append(i.find_dist(test_pt))

# test_pt_plot = np.zeros()
plot_dist(x1,y1,z1+z2,flag=True,test_point=test_pt)
    
print(distances)
print(f"Classified as class {np.argmin(distances)}")
```
